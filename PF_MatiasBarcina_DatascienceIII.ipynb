```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
from wordcloud import WordCloud  # Importar WordCloud
import matplotlib.pyplot as plt  # Importar matplotlib para visualización

# Descargar recursos de NLTK (solo la primera vez)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')  # Asegurar que este recurso esté disponible

# Cargar el archivo de texto
with open('Don Quijote.txt', 'r', encoding='utf-8') as file:
    texto = file.read()

# Tokenización: Dividir el texto en palabras
tokens = word_tokenize(texto.lower())  # Convertir a minúsculas y tokenizar

# Limpieza: Eliminar caracteres especiales y números
tokens = [word for word in tokens if word.isalnum()]

# Eliminar stopwords (palabras irrelevantes)
stop_words = set(stopwords.words('spanish'))  # Cambia a 'english' si el texto está en inglés
tokens = [word for word in tokens if word not in stop_words]

# Lematización: Reducir palabras a su forma base
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(word) for word in tokens]

# Frecuencia de palabras
frecuencia = Counter(tokens)
print("Palabras más comunes:", frecuencia.most_common(10))

# Nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(frecuencia)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Cargar el dataset MNIST
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalizar los datos (escalar los valores de píxeles al rango [0, 1])
X_train = X_train / 255.0
X_test = X_test / 255.0

# Construir la red neuronal
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Aplanar la imagen de 28x28 a un vector de 784 elementos
    layers.Dense(128, activation='relu'),  # Capa oculta con 128 neuronas y activación ReLU
    layers.Dense(10, activation='softmax') # Capa de salida con 10 neuronas (una por cada clase) y activación softmax
])

# Compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entrenar el modelo
print("Entrenando la red neuronal...")
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Evaluar el modelo
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Precisión en el conjunto de prueba: {test_acc:.4f}")

# Graficar la precisión y la pérdida durante el entrenamiento
plt.figure(figsize=(12, 4))

# Gráfico de precisión
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Precisión en entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión en validación')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.title('Precisión durante el entrenamiento')
plt.legend()

# Gráfico de pérdida
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Pérdida en entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida en validación')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.title('Pérdida durante el entrenamiento')
plt.legend()

plt.show()
```

Precisión en entrenamiento (accuracy):

Comenzó en 88.02% en la primera época y aumentó hasta 98.67% en la quinta época.

Esto indica que el modelo está aprendiendo correctamente a clasificar las imágenes del conjunto de entrenamiento.

Pérdida en entrenamiento (loss):

La pérdida disminuyó de 0.4301 a 0.0441.

Una pérdida más baja significa que el modelo está haciendo predicciones más precisas.

Precisión en validación (val_accuracy):

Comenzó en 95.91% y aumentó hasta 97.72%.

Esto indica que el modelo generaliza bien a datos no vistos (conjunto de prueba).

Pérdida en validación (val_loss):

La pérdida en validación disminuyó de 0.1376 a 0.0748.

Esto confirma que el modelo no está sobreajustando (overfitting) los datos de entrenamiento.

Evaluación final:

En el conjunto de prueba, el modelo alcanzó una precisión de 97.35% con una pérdida de 0.0863.

Esto significa que el modelo clasifica correctamente el 97.35% de las imágenes de prueba.

Gráficos de precisión y pérdida
Los gráficos que se generaron muestran:

Precisión: Cómo aumenta la precisión en entrenamiento y validación a medida que avanza el entrenamiento.

Pérdida: Cómo disminuye la pérdida en entrenamiento y validación.

Estos gráficos son útiles para detectar problemas como el sobreajuste (overfitting) o el subajuste (underfitting). En tu caso, ambos gráficos muestran un comportamiento saludable, ya que la precisión aumenta y la pérdida disminuye de manera consistente.


```python
from nltk.sentiment import SentimentIntensityAnalyzer

# Descargar recurso de VADER (solo la primera vez)
nltk.download('vader_lexicon')

# Crear el analizador de sentimientos
analyzer = SentimentIntensityAnalyzer()

# Analizar el sentimiento de un texto
texto = "Este es un texto de ejemplo. ¡Me encanta aprender NLP!"
sentimiento = analyzer.polarity_scores(texto)
print(sentimiento)
```

El resultado que obtuviste, {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}, es el análisis de sentimientos realizado por VADER (Valence Aware Dictionary and sEntiment Reasoner). VADER es una herramienta de análisis de sentimientos que está especialmente diseñada para textos en inglés y utiliza un léxico de palabras con puntuaciones de sentimiento predefinidas.

Interpretación del resultado
neg: Puntuación de sentimiento negativo (en este caso, 0.0).

neu: Puntuación de sentimiento neutral (en este caso, 1.0).

pos: Puntuación de sentimiento positivo (en este caso, 0.0).

compound: Puntuación compuesta que resume el sentimiento general (en este caso, 0.0).

En este caso, el texto analizado se clasificó como completamente neutral (neu: 1.0), lo que significa que no se detectaron palabras con carga emocional (ni positiva ni negativa).
